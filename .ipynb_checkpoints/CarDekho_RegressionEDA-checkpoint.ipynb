```python
"""This dataset contains information about used cars listed on www.cardekho.com.
Problem Statement:
You need to do a detailed analysis (both EDA and statistical analysis) on the cars dataset.
You need to build a Machine Learning model to predict the selling_price.

Expectations:
Since it is a regression based project, you are expected to build a Linear Regression model
with all the performance metrics like R-squared, Adjusted R-squared, MAE, MSE, RMSE,RSS.
"""
#====== Importing the required Library===============
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from  sklearn.preprocessing import minmax_scale
import random
#====================================================
# Import the data set into workspace:
car = pd.read_csv("D:\Learn & Projects\Data Science\Datasets\CarDekho\carDetails.csv")    
```

```python
#Summary
car.info(verbose=True) 
#Missing value - column wise
print(pd.isnull(car).sum())
print(car.isna().sum())
new_colName = ['Year', 'SellingPrice', 'KmDriven', 'Fuel', 'SellerType', 'Transmission', 'Owner', 'Make', 'Model']   

```

```python
# Feature Trasnformation : Separating Name column in Make and Model column
car[['Make','Model']] = car.name.apply(lambda x : pd.Series(str(x).split(" ",1)))

car_dekho = car.loc[:,car.columns!='name']

#car_dekho=car_dekho.set_axis(['Year', 'SellingPrice', 'KmDriven', 'Fuel', 'SellerType', 'Transmission', 'Owner', 'Make', 'Model'], axis=1)

car_dekho = car_dekho.rename(columns=lambda x, *, y=iter(new_colName): next(y))
car_dekho.head()
```

```python
#Store a list of columns in variable:
colName = list(car_dekho.columns)    
print(colName)    
car_dekho['Make'].value_counts().plot(kind='bar') 
tempMakedf= car_dekho[colName[7]].value_counts().rename_axis('Make').to_frame('Counts').reset_index()
```

```python
"""Function for creating bar graph between 2 variables and adding labels on top of bars"""
def labels_barPlot(dataFrame_Name,xAxisColName,yAxisColName,xlabelName,ylabelName,
                   grphTitle,extrahtforlabel =5,sizelabel=13,
                  figureht=8,figurewidt=12,xlabelNameSize=18,ylabelNamesize=18,grphTitlesize=18):
    
    plt.figure(figsize=(figurewidt, figureht))
    plots= sns.barplot(x=xAxisColName,y=yAxisColName,data=dataFrame_Name)
    """
  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # sfree space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    """
    for bar in plots.patches:
        plots.annotate(format(bar.get_height(), '.0f'),(bar.get_x() + bar.get_width() / 2, 
                                                        bar.get_height()+extrahtforlabel), ha='center', 
                       va='center', size=sizelabel, xytext=(0, 8),textcoords='offset points')   
        
    
    # Setting the label for x-axis
    plt.xlabel(xlabelName, size=xlabelNameSize)
      
    # Setting the label for y-axis
    plt.ylabel(ylabelName, size=ylabelNamesize)
      
    # Setting the title for the graph
    plt.title(grphTitle, size=grphTitlesize)
    # Fianlly showing the plot
    plt.tight_layout()
   # plt.show()   
#------------------------------------------------------------------------------------------    
```

```python
#--- How many kms was driven for different type of transmission car
trans_km =car_dekho.groupby(['Transmission']).sum()['KmDriven'].rename_axis('Trans').to_frame('TotalKM').reset_index()

labels_barPlot(trans_km,'Trans','TotalKM',
               'Type of Transmission',
               'Total KM driven','Transmission wise km driven')
```

```python
fuelType_kmDriven_carMake = car_dekho.groupby(['Fuel','Make'])['KmDriven'].count().reset_index()

for i in fuelType_kmDriven_carMake['Fuel'].unique():
    temp_df=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake['Fuel']==i]
    for j in fuelType_kmDriven_carMake['Make'].unique():
            if temp_df[temp_df['Make']==j].shape[0]==0:
                fuelType_kmDriven_carMake=pd.concat([fuelType_kmDriven_carMake,pd.DataFrame({'Fuel':[i],'Make':[j],'KmDriven':[0]})],axis=0)
               
               
petrol=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake["Fuel"]=="Petrol"]["KmDriven"]
diesel=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake["Fuel"]=="Diesel"]["KmDriven"]
cng=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake["Fuel"]=="CNG"]["KmDriven"]
lpg=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake["Fuel"]=="LPG"]["KmDriven"]
electric=fuelType_kmDriven_carMake[fuelType_kmDriven_carMake["Fuel"]=="Electric"]["KmDriven"]



fig,ax = plt.subplots() #Returns tuple
ax.bar(fuelType_kmDriven_carMake['Make'].unique(),petrol,label="Petrol")
ax.bar(fuelType_kmDriven_carMake['Make'].unique(),diesel,label="Diesel")
ax.bar(fuelType_kmDriven_carMake['Make'].unique(),cng,label="CNG")
ax.bar(fuelType_kmDriven_carMake['Make'].unique(),lpg,label="LPG")
ax.bar(fuelType_kmDriven_carMake['Make'].unique(),electric,label="Electric")
ax.set_xlabel("Cars")
ax.set_ylabel("Total kms driven")
ax.set_title("Cars and Fuel type wise total kms driven")
ax.legend()
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')
plt.show()
```

```python
# Selling Price vs Transmission
plt.figure(figsize=(12, 8))

sns.boxplot(x="Transmission", y='SellingPrice',data=car_dekho, linewidth=2.5)
# Setting the label for x-axis
plt.xlabel("Year", size=15)

# Setting the label for y-axis
plt.ylabel("Selling Price", size=15)

# Setting the title for the graph
plt.title("Scatter Plot between year and Selling Price", size=15)
# Fianlly showing the plot
plt.tight_layout()
```

```python
#--------------------Count of seller type year wise------------------------------------------
#Year wise total number of car sold by seller type
year_sellerType_count = car_dekho.groupby(['SellerType','Year'])['KmDriven'].count().reset_index()
year_sellerType_count.head()

year_sellerType_count.rename(columns={'KmDriven':'CountOfCarsSold'},inplace=True)
year_sellerType_count.head()
```

```python
for i in year_sellerType_count['SellerType'].unique():
    temp_cardf = year_sellerType_count[year_sellerType_count['SellerType']==i]
    for j in year_sellerType_count['Year'].unique():
        if temp_cardf[temp_cardf['Year']==j].shape[0]==0:
            year_sellerType_count=pd.concat([year_sellerType_count,pd.DataFrame({'SellerType':[i],'Year':[j],'CountOfCarsSold':[0]})],
                                            axis=0)
individual = year_sellerType_count[year_sellerType_count['SellerType']=='Individual']['CountOfCarsSold']
dealer = year_sellerType_count[year_sellerType_count['SellerType']=='Dealer']['CountOfCarsSold']
trademark_dealer = year_sellerType_count[year_sellerType_count['SellerType']=='Trustmark Dealer']['CountOfCarsSold']
```

```python
# X axis as Year

fig,ax = plt.subplots(1,figsize=(16, 10)) #Returns tuple
ax.bar(year_sellerType_count['Year'].unique(),individual,label="Individual")
ax.bar(year_sellerType_count['Year'].unique(),dealer,label="Dealer")
ax.bar(year_sellerType_count['Year'].unique(),trademark_dealer,label="Trustmark Dealer")
ax.legend()
ax.set_xlabel("Year")
ax.set_ylabel("Number of cars Sold")
ax.set_title("Yearwise cars sold by seller type")
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')
plt.show() 
```

```python
# X axis as Seller type
fig,ax = plt.subplots(1,figsize=(12, 10)) #Returns tuple

for i in year_sellerType_count['Year'].unique():
    year_x = year_sellerType_count[year_sellerType_count['Year']==i]
    ax.bar(year_x['SellerType'].unique(),year_x['CountOfCarsSold'],label="Individual")
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical') 
plt.legend(list(year_sellerType_count['Year'].unique()), bbox_to_anchor=([0.22, 1, 0, 0]), ncol=2, frameon=True)
ax.set_xlabel("Seller Type")
ax.set_ylabel("Number of cars Sold")
ax.set_title("Yearwise cars sold by seller type")
plt.show()
#=====================================================================
```
